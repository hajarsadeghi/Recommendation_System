```python
#Creating A Function to extract features

def feature_extractor(adf, group, save_dst = dst_features, train_data_dir = '/content', image_width = 224, image_height = 224, model = model):
  t1 = time()
  print('Extracting Features For Group:',group)
  n_train_samples = adf.shape[0]
  art_ids = []
  datagen = ImageDataGenerator(rescale=1. / 255)
  generator = datagen.flow_from_directory(
      train_data_dir,
      target_size=(image_width, image_height),
      batch_size=1,
      # class_mode=None,
      classes = [group],
      shuffle=False)
  for fp in generator.filenames:
      art_ids.append(fp[(fp.find("/")+2):i.find(".")])

  # print(art_ids)
  art_features = model.predict_generator(generator, n_train_samples) #Batch Size = 1
  # art_features = model.predict(generator, steps = n_train_samples, batch_size=1) #Batch Size = 1
  print(art_features.shape)
  art_features = art_features.reshape((art_features.shape[0], np.prod(art_features.shape[1:])))
  print(art_features.shape)

  print('Time Taken To Extract Features = ',time()-t1,'seconds')

  #Save The Features
  np.save(open(save_dst+'/'+group+'_ResNet_features.npy', 'wb'), art_features)
  np.save(open(save_dst+'/'+group+'_ResNet_feature_product_ids.npy', 'wb'), np.array(art_ids))
```
